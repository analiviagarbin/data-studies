Pagination

```python
api_base_url = 'https://api.github.com'
owner = 'amzn' # login/username
url = f'{api_base_url}/users/{owner}/repos'

url
```

```python
repos_list = []
page = 1

while True:
    try:
        url_page = f'{url}?page={page}'
        response = requests.get(url_page, headers=headers)
        
        if len(response.json()) == 0:
            break
        
        repos_list.append(response.json())
        page += 1
    except:
        repos_list.append(None)
```

```python
repos_list
```

```python
# followers data

headers = {'Authorization': 'Bearer ' + access_token,
           'X-GitHub-Api-Version': '2022-11-28'}

api_base_url = 'https://api.github.com'
owner = 'amzn' # login/username

f_url = f'{api_base_url}/users/{owner}/followers'

followers_list = []
i = 1

while True:
    try:
        url_page = f'{f_url}?page={i}'
        f_response = requests.get(url_page, headers=headers)
        followers = f_response.json()

        if len(followers) == 0:
            break

        followers_list.append(followers)
        i += 1
    except:
        followers_list.append(None)
        break
```

```python
followers_list
```

Transforming data

Repo names

```python
repos_name = []

for page in repos_list:
    for repo in page:
        repos_name.append(repo['name'])

repos_name[:10]
```

```python
len(repos_name)
```

Repos language

```python
repos_lang = []

for page in repos_list:
    for repo in page:
        repos_lang.append(repo['language'])

repos_lang[:10]
```

```python
len(repos_lang)
```

DataFrame

```python
import numpy as np
import pandas as pd
```

```python
data_amz = pd.DataFrame()
data_amz['repository_name'] = repos_name
data_amz['language'] = repos_lang
```
